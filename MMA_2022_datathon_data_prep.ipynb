## Break down by department, see number of transactions by manufacturer by each department to see which manufacturers are most competitive in each department

```python
dept_manu_trans_rev_df = transaction_df.groupby(['DEPARTMENT','MANUFACTURER']).agg(num_transactions=('SALES_VALUE', 'count'), total_revenue=('SALES_VALUE', 'sum')).sort_values(by = ['num_transactions', 'total_revenue'], ascending = [False, False])
```

```python
dept_manu_trans_rev_df
```

```python
dept_manu_trans_rev_df = dept_manu_trans_rev_df.reset_index()
```

```python
# getting top 10 manufacturers from each department based on highest rev and highest transaction numbers
department_list = set(dept_manu_trans_rev_df['DEPARTMENT'])
cols = ['Department', 'Top_manufacturers']
dept_manu_df = pd.DataFrame(columns = cols)
all_top_10s = []
for dept in department_list:
  top_10_manu = dept_manu_trans_rev_df.loc[dept_manu_trans_rev_df.DEPARTMENT == dept, 'MANUFACTURER'].value_counts().head(10).index.values.tolist() # data is already sorted for top manufacturers only so no need here again
  dept_manu_df = dept_manu_df.append({'Department': dept , 'Top_manufacturers':top_10_manu},ignore_index=True)  
  all_top_10s.append(top_10_manu)
dept_manu_dict = dict(zip(department_list, all_top_10s))
```

```python
dept_manu_df
```

```python
# dictionary format of the same dataframe above - 
dept_manu_dict
```

```python
# Departments of all top 10 Manufacturers
all_manu = set(np.concatenate([x for x in dept_manu_dict.values()]))
all_manu_df = dept_manu_trans_rev_df[dept_manu_trans_rev_df.MANUFACTURER.isin(all_manu)]
all_manu_df = all_manu_df.groupby(by = 'MANUFACTURER')['DEPARTMENT'].apply(list).reset_index()
all_manu_df['Total_Dept'] = all_manu_df.apply(lambda x: len(x.DEPARTMENT), axis =1)
all_manu_df = all_manu_df.sort_values(by = 'Total_Dept', ascending = False)
all_manu_df.head()
```

```python
##Adding total Sales and Discount for Each of the TOP 10 Manufacturer
all_manu_transaction = transaction_df[transaction_df.MANUFACTURER.isin(all_manu)]
all_manu_transaction = all_manu_transaction.groupby('MANUFACTURER').agg(SALES = ('SALES_VALUE','sum'), QUANTITY = ('QUANTITY','sum'),Discount_amt = ('discount_amt','sum'),COUPON_MATCH_DISC = ('COUPON_MATCH_DISC','sum'))
all_manu_transaction.COUPON_MATCH_DISC = abs(all_manu_transaction.COUPON_MATCH_DISC)
all_manu_transaction = all_manu_df.merge(all_manu_transaction,on ='MANUFACTURER', how ='inner')
all_manu_transaction.head(30)
```

```python
all_manu_transaction.to_csv('all_manu_transaction.csv')
```

```python
# top_10_man_exploded
```

```python
### Dividing quanity SALES and Discount for each Department
top_10_man = all_manu_transaction
top_10_man_exploded = top_10_man.explode("DEPARTMENT")
top_10_man_merged = top_10_man_exploded.drop(columns=["QUANTITY","COUPON_MATCH_DISC"]).merge(transaction_df, on=["MANUFACTURER", "DEPARTMENT"]).loc[:, ["MANUFACTURER", "DEPARTMENT", "SALES_VALUE", "QUANTITY", "discount_amt","COUPON_MATCH_DISC"]]
top_10_man_merged = top_10_man_merged.groupby(by=["MANUFACTURER", "DEPARTMENT"]).agg(Sales_sum = ('SALES_VALUE','sum'), Quantity_sum = ('QUANTITY','sum'), Discount_amt_median = ('discount_amt','median'),Total_Disc = ('discount_amt','sum'),COUPON_MATCH_DISC_SUM = ('COUPON_MATCH_DISC','sum') )
top_10_man_merged.COUPON_MATCH_DISC_SUM = abs(top_10_man_merged.COUPON_MATCH_DISC_SUM)
top_10_man_merged = top_10_man_merged.assign(Original_Price=lambda x: (x['Sales_sum']+(x['Total_Disc'])/x['Quantity_sum']))
top_10_man_merged = top_10_man_merged.assign(Percentage_Discount = lambda x: ((x['Original_Price'] - (x['Sales_sum']))/x['Original_Price']))
top_10_man_merged = top_10_man_merged.assign(Original_Price_Manu=lambda x: (x['Sales_sum']+(x['COUPON_MATCH_DISC_SUM'])/x['Quantity_sum']))
top_10_man_merged = top_10_man_merged.assign(Percentage_Discount_Manu = lambda x: ((x['Original_Price_Manu'] - (x['Sales_sum']))/x['Original_Price_Manu']))
top_10 = top_10_man_merged.copy()
top_10_man_merged.to_csv('top_10_man_merged.csv')
top_10_man_merged
```

```python
# creating function 

def overall_model(df, groupby_cols :list, pivot_val : int):
  top_10_man_merged_perc_mat = df.groupby(groupby_cols).agg(Sales_sum = ('Sales_sum','sum'),Quantity_sum = ('Quantity_sum','sum'),
                                                                                                     Total_Disc = ('Total_Disc','sum'),COUPON_MATCH_DISC_SUM = ('COUPON_MATCH_DISC_SUM','sum')
                                                                                                     ,Orignal_Price = ('Original_Price','sum'),Orignal_Price_Manu = ('Original_Price_Manu','sum'))
  top_10_man_merged_perc_mat = top_10_man_merged_perc_mat.assign(Percentage_Discount_Manu = lambda x: ((x['Orignal_Price_Manu'] - (x['Sales_sum']))/x['Orignal_Price_Manu']))
  top_10_man_merged_perc_mat = pd.pivot_table(top_10_man_merged_perc_mat,index = groupby_cols[0],columns = groupby_cols[1],values = pivot_val,aggfunc = sum, fill_value=0)
  top_10_man_merged_perc_mat_1 = top_10_man_merged_perc_mat
  for column in top_10_man_merged_perc_mat.columns[::-1]:
    top_10_man_merged_perc_mat_1[column] = top_10_man_merged_perc_mat[column].rank(pct=True)#Changining into perc
  top_10_man_merged_perc_mat_2 = top_10_man_merged_perc_mat_1.copy()
  for column in top_10_man_merged_perc_mat_2.columns[::-1]:
    top_10_man_merged_perc_mat_2.loc[top_10_man_merged_perc_mat_2[column] >= .80, column] = 1#Setting Percentile Constraint
    top_10_man_merged_perc_mat_2.loc[top_10_man_merged_perc_mat_2[column] < .80, column] = 0
  top_10_man_merged_perc_mat_2[f'{pivot_val}_Total'] = top_10_man_merged_perc_mat_2[top_10_man_merged_perc_mat_2.columns].sum(axis = 1) # Adding number of departments a product is top in
  top_10_man_merged_perc_mat_2 = top_10_man_merged_perc_mat_2.reset_index()
  manufacturers, totals = top_10_man_merged_perc_mat_2['MANUFACTURER'], top_10_man_merged_perc_mat_2[f'{pivot_val}_Total']
  resdf = pd.DataFrame(list(zip(manufacturers, totals)),columns=['MANUFACTURER',f'{pivot_val}_Total'])
  return resdf
```
